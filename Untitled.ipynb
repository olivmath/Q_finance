```python
from requests import get, post
import pandas as pd
import numpy as np
```

```python
url = 'https://api3.binance.com'
klines = url + '/api/v3/klines'
info = url + '/api/v3/exchangeInfo'
```

```python
r_json = get(info).json()
```

```python
for i in r_json['symbols']:
    if "BTCUSDT" in i['symbol']: print(i)
```

```python
params = { 
    "symbol": "BTCUSDT",
    "interval": "1d",
    "limit": 1000
}
```

```python
r = get(klines, params=params)
```

```python
r_json = r.json()
```

```python
df = pd.DataFrame(r_json, columns=['Open time',
                                   'Open',
                                   'High',
                                   'Low',
                                   'Close',
                                   'Volume',
                                   'Close time',
                                   'Quote asset volume',
                                   'Number of trades',
                                   'Taker buy base asset volume',
                                   'Taker buy quote asset volume',
                                   'Ignore'
                                   ])
```
